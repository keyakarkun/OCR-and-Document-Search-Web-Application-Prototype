```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import numpy as np

# Read image
image_path = 'C:/Users/keyak/OneDrive/Desktop/OCR/images/hindi.png'

img = cv2.imread(image_path)

# Instance text detector for both English and Hindi languages
reader = easyocr.Reader(['en', 'hi'], gpu=False)

# Detect text on image
text_ = reader.readtext(img)

threshold = 0.25

# Draw bbox and text
for t_, t in enumerate(text_):
    print(t)

    bbox, text, score = t

    if score > threshold:
        # Convert bbox points to integers
        x1, y1 = int(bbox[0][0]), int(bbox[0][1])  # Convert to int
        x2, y2 = int(bbox[2][0]), int(bbox[2][1])  # Convert to int
        
        # Draw rectangle and put text
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 5)
        cv2.putText(img, text, (x1, y1 - 10), cv2.FONT_HERSHEY_COMPLEX, 0.65, (255, 0, 0), 2)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
